```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import joblib
from sklearn.preprocessing import LabelEncoder, MinMaxScaler
from sklearn.model_selection import train_test_split
import tensorflow as tf
from tensorflow.keras.preprocessing.image import load_img, img_to_array
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.layers import Input, Dense, Dropout, GlobalAveragePooling2D, concatenate, BatchNormalization
from tensorflow.keras.models import Model
from tensorflow.keras.applications import EfficientNetB0, MobileNetV2
from tensorflow.keras.optimizers import Adam
```

```python
# -------------- SETTINGS --------------
IMAGE_SIZE = 224         # 224 for EfficientNet/MobileNet; use 128 to speed up
BATCH_SIZE = 16          # reduce for low-RAM machines
EPOCHS = 15              # change as needed
USE_EFFICIENTNET = False # set False to use MobileNetV2 (lighter on CPU)
DATA_DIR = '.'           # your working folder
IMG_FOLDER_1 = os.path.join(DATA_DIR, 'HAM10000_images_part_1')
IMG_FOLDER_2 = os.path.join(DATA_DIR, 'HAM10000_images_part_2')
META_CSV = os.path.join(DATA_DIR, 'HAM10000_metadata.csv')
CHARTS_DIR = os.path.join(DATA_DIR, 'charts')
os.makedirs(CHARTS_DIR, exist_ok=True)
```

```python
# -------------- 1. LOAD METADATA --------------
meta = pd.read_csv(META_CSV)
print("Metadata loaded:", meta.shape)
# ensure image path column
def get_image_path(image_id):
    p1 = os.path.join(IMG_FOLDER_1, image_id + '.jpg')
    p2 = os.path.join(IMG_FOLDER_2, image_id + '.jpg')
    if os.path.exists(p1): return p1
    if os.path.exists(p2): return p2
    return None

meta['image_path'] = meta['image_id'].apply(get_image_path)
meta = meta.dropna(subset=['image_path']).reset_index(drop=True)
print("After filtering missing images:", meta.shape)
```

```python
# 1. Set paths
# ------------------------------
DATA_DIR = '.'  # change to your working directory
META_CSV = os.path.join(DATA_DIR, 'HAM10000_metadata.csv')
# 2. Load the dataset
# ------------------------------
metadata = pd.read_csv(META_CSV)
print("Metadata loaded successfully!")
print("Total records:", metadata.shape[0])
# 3. Show first 5 rows
# ------------------------------
print("\nFirst 5 rows of the dataset:")
print(metadata.head())
```

```python
# -------------- 2. EDA (simple plots for project report) --------------
# class counts
plt.figure(figsize=(8,4))
meta['dx'].value_counts().plot(kind='bar')
plt.title('Class counts (dx)')
plt.tight_layout()
plt.savefig(os.path.join(CHARTS_DIR, 'class_counts_dx.png'))
plt.close()

# age distribution
plt.figure(figsize=(8,4))
meta['age'].dropna().astype(float).plot(kind='hist', bins=30)
plt.title('Age distribution')
plt.tight_layout()
plt.savefig(os.path.join(CHARTS_DIR, 'age_distribution.png'))
plt.close()

# sex vs dx
plt.figure(figsize=(8,4))
pd.crosstab(meta['dx'], meta['sex']).plot(kind='bar', stacked=True, figsize=(10,4))
plt.title('Sex distribution by dx')
plt.tight_layout()
plt.savefig(os.path.join(CHARTS_DIR, 'sex_by_dx.png'))
plt.close()
```

```python
# -------------- 3. PREPROCESS METADATA --------------
# fill age
meta['age'] = meta['age'].fillna(meta['age'].median())

# normalize localization strings (some cleaning)
meta['localization'] = meta['localization'].astype(str).str.lower().str.replace('[^a-z0-9 _-]', '', regex=True)

# label encode categorical fields
le_sex = LabelEncoder()
meta['sex_enc'] = le_sex.fit_transform(meta['sex'].astype(str))

le_loc = LabelEncoder()
meta['loc_enc'] = le_loc.fit_transform(meta['localization'].astype(str))

# encode dx (7 classes)
le_dx = LabelEncoder()
meta['dx_enc'] = le_dx.fit_transform(meta['dx'].astype(str))
class_names = list(le_dx.classes_)
print("Class names (dx):", class_names)

# save encoders so app can use them later
joblib.dump(le_sex, os.path.join(DATA_DIR, 'le_sex.pkl'))
joblib.dump(le_loc, os.path.join(DATA_DIR, 'le_loc.pkl'))
joblib.dump(le_dx, os.path.join(DATA_DIR, 'le_dx.pkl'))

# prepare tabular features
tabular = meta[['age','sex_enc','loc_enc']].values
scaler = MinMaxScaler()
tabular_scaled = scaler.fit_transform(tabular)
joblib.dump(scaler, os.path.join(DATA_DIR, 'scaler_tabular.pkl'))
```

```python
# -------------- 4. IMAGE LOADER (careful with memory) --------------
# We'll build tf.data pipeline to avoid loading all images at once
def preprocess_image_file(path):
    img = load_img(path, target_size=(IMAGE_SIZE, IMAGE_SIZE))
    arr = img_to_array(img).astype('float32') / 255.0
    return arr

# Build arrays of paths, tabular, labels
paths = meta['image_path'].values
tabs = tabular_scaled
labels = to_categorical(meta['dx_enc'], num_classes=len(class_names))

# train/test split (stratify on dx_enc)
train_idx, test_idx = train_test_split(
    np.arange(len(paths)), test_size=0.2, random_state=42, stratify=meta['dx_enc']
)

paths_train, paths_test = paths[train_idx], paths[test_idx]
tabs_train, tabs_test = tabs[train_idx], tabs[test_idx]
y_train, y_test = labels[train_idx], labels[test_idx]

print("Train size:", len(paths_train), "Test size:", len(paths_test))

# create tf.data dataset
def gen(paths_array, tabs_array, labels_array, batch_size=BATCH_SIZE, shuffle=True):
    dataset = tf.data.Dataset.from_tensor_slices((paths_array, tabs_array, labels_array))
    if shuffle:
        dataset = dataset.shuffle(buffer_size=len(paths_array), reshuffle_each_iteration=True)
    def _read(path, tab, label):
        img = tf.io.read_file(path)
        img = tf.image.decode_jpeg(img, channels=3)
        img = tf.image.resize(img, [IMAGE_SIZE, IMAGE_SIZE])
        img = img / 255.0
        return (img, tab), label
    dataset = dataset.map(_read, num_parallel_calls=tf.data.AUTOTUNE)
    dataset = dataset.batch(batch_size).prefetch(tf.data.AUTOTUNE)
    return dataset

train_ds = gen(paths_train, tabs_train, y_train, batch_size=BATCH_SIZE, shuffle=True)
val_ds = gen(paths_test, tabs_test, y_test, batch_size=BATCH_SIZE, shuffle=False)
```

```python
# -------------- 5. BUILD MULTIMODAL MODEL --------------
# Image branch
img_input = Input(shape=(IMAGE_SIZE, IMAGE_SIZE, 3), name='img_input')
if USE_EFFICIENTNET:
    base = EfficientNetB0(weights='imagenet', include_top=False, input_tensor=img_input)
else:
    base = MobileNetV2(weights='imagenet', include_top=False, input_tensor=img_input, alpha=1.0)

x = base.output
x = GlobalAveragePooling2D()(x)
x = Dense(128, activation='relu')(x)
x = BatchNormalization()(x)
x = Dropout(0.3)(x)

# Tabular branch
tab_input = Input(shape=(tabs_train.shape[1],), name='tab_input')
t = Dense(32, activation='relu')(tab_input)
t = Dropout(0.2)(t)
t = Dense(16, activation='relu')(t)

# Merge
merged = concatenate([x, t])
z = Dense(128, activation='relu')(merged)
z = Dropout(0.4)(z)
output = Dense(len(class_names), activation='softmax', name='output')(z)

model = Model(inputs=[img_input, tab_input], outputs=output)

# freeze base model optionally (we'll keep it frozen for initial training)
for layer in base.layers:
    layer.trainable = False

model.compile(optimizer=Adam(1e-4), loss='categorical_crossentropy', metrics=['accuracy'])
model.summary()
```

```python
# -------------- 6. TRAIN --------------
history = model.fit(train_ds, validation_data=val_ds, epochs=EPOCHS)
```

```python
import os
import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow.keras.preprocessing import image
from tensorflow.keras.models import load_model

# ------------------------------
# 1. Load your trained multiclass model
# ------------------------------
model_path = 'multiclass_skin_cancer.keras'  # your trained multiclass model
model = load_model(model_path)
print("Multiclass model loaded successfully!")

# ------------------------------
# 2. Load metadata
# ------------------------------
metadata_path = 'HAM10000_metadata.csv'  # metadata with labels
metadata = pd.read_csv(metadata_path)

# ------------------------------
# 3. Define class mapping for HAM10000
# ------------------------------
class_mapping = {
    'nv': 0,       # melanocytic nevi
    'mel': 1,      # melanoma
    'bkl': 2,      # benign keratosis-like lesions
    'bcc': 3,      # basal cell carcinoma
    'akiec': 4,    # actinic keratoses
    'vasc': 5,     # vascular lesions
    'df': 6        # dermatofibroma
}

# Reverse mapping to get class names from predictions
class_names = {v:k for k,v in class_mapping.items()}

# ------------------------------
# 4. Function to predict a single image
# ------------------------------
def predict_image(img_path):
    # Load and preprocess image
    img = image.load_img(img_path, target_size=(224, 224))
    img_array = image.img_to_array(img) / 255.0
    img_array = np.expand_dims(img_array, axis=0)  # add batch dimension

    # Predict
    preds = model.predict(img_array)
    pred_class = np.argmax(preds, axis=1)[0]
    confidence = preds[0][pred_class]

    # Map predicted class index to label
    pred_label = class_names[pred_class]

    print(f"Image: {os.path.basename(img_path)}")
    print(f"Predicted Class: {pred_label}")
    print(f"Confidence: {confidence*100:.2f}%\n")
    return pred_label, confidence

# ------------------------------
# 5. Test multiple images
# ------------------------------
# Example: test images from your folders
test_folders = ['HAM10000_images_part_1', 'HAM10000_images_part_2']

# Loop through first 10 images from each folder
for folder in test_folders:
    folder_path = os.path.join('.', folder)
    images = os.listdir(folder_path)[:10]  # first 10 images
    for img_name in images:
        img_path = os.path.join(folder_path, img_name)
        predict_image(img_path)
```

```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras.preprocessing import image
from tensorflow.keras.models import load_model

# ------------------------------
# 1. Load trained model
# ------------------------------
model_path = 'multiclass_skin_cancer.keras'
model = load_model(model_path)
print("Model loaded successfully!")

# ------------------------------
# 2. Load metadata
# ------------------------------
metadata_path = 'HAM10000_metadata.csv'
metadata = pd.read_csv(metadata_path)

# ------------------------------
# 3. Define class mapping
# ------------------------------
class_mapping = {
    'nv': 0,
    'mel': 1,
    'bkl': 2,
    'bcc': 3,
    'akiec': 4,
    'vasc': 5,
    'df': 6
}
inv_class_mapping = {v: k for k, v in class_mapping.items()}

# ------------------------------
# 4. Prepare images for grid display
# ------------------------------
image_folders = ['HAM10000_images_part_1', 'HAM10000_images_part_2']

# Collect images, actual labels, predictions, confidence
images = []
actual_labels = []
pred_labels = []
confidences = []

num_images_to_show = 16  # total images in grid
count = 0

for folder in image_folders:
    folder_path = os.path.join('.', folder)
    img_names = os.listdir(folder_path)
    
    for img_name in img_names:
        img_path = os.path.join(folder_path, img_name)
        
        # Load and preprocess image
        img = image.load_img(img_path, target_size=(224, 224))
        img_array = image.img_to_array(img) / 255.0
        img_array_exp = np.expand_dims(img_array, axis=0)
        
        # Predict
        preds = model.predict(img_array_exp)
        pred_class = np.argmax(preds, axis=1)[0]
        confidence = np.max(preds)
        
        # Actual label
        img_id = img_name.split('.')[0]
        label_row = metadata[metadata['image_id'] == img_id]
        if not label_row.empty:
            actual_label = label_row['dx'].values[0]
        else:
            actual_label = "Unknown"
        
        # Save info for grid
        images.append(img)
        actual_labels.append(actual_label)
        pred_labels.append(inv_class_mapping.get(pred_class, 'Unknown'))
        confidences.append(confidence)
        
        count += 1
        if count >= num_images_to_show:
            break
    if count >= num_images_to_show:
        break

# ------------------------------
# 5. Display images in grid
# ------------------------------
grid_size = int(np.ceil(np.sqrt(num_images_to_show)))
plt.figure(figsize=(12, 12))

for i in range(num_images_to_show):
    plt.subplot(grid_size, grid_size, i + 1)
    plt.imshow(images[i])
    plt.axis('off')
    plt.title(f"A: {actual_labels[i]}\nP: {pred_labels[i]}\n{confidences[i]*100:.1f}%")

plt.tight_layout()
plt.show()
```

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.preprocessing import image
from tensorflow.keras.models import load_model

# 1. Load model
model_path = 'multiclass_skin_cancer.keras'
model = load_model(model_path)
print("Model loaded successfully!")

# 2. Manual feature mappings
sex_mapping = {'male':0, 'female':1}
loc_mapping = {
    'abdomen':0,'acral':1,'back':2,'chest':3,'ear':4,'face':5,'foot':6,'genital':7,
    'hand':8,'lower extremity':9,'neck':10,'scalp':11,'trunk':12,'unknown':13,'upper extremity':14
}

# 3. User input
age = float(input("Enter age: "))
sex = input("Enter sex (male/female): ").lower()
location = input("Enter lesion location: ").lower()

sex_encoded = sex_mapping.get(sex, 0)
loc_encoded = loc_mapping.get(location, 13)
age_scaled = age / 100.0
tabular_features = np.array([[age_scaled, sex_encoded, loc_encoded]])

# 4. Image path input
image_path = input("Enter full path to the lesion image: ")

# 5. Load & preprocess image
img = image.load_img(image_path, target_size=(224,224))
img_array = image.img_to_array(img)/255.0
img_array_exp = np.expand_dims(img_array, axis=0)

# 6. Predict
class_names = ['akiec','bcc','bkl','df','mel','nv','vasc']
preds = model.predict(img_array_exp)
pred_class_idx = np.argmax(preds)
confidence = np.max(preds)
pred_class = class_names[pred_class_idx]

# 7. Show result
plt.figure(figsize=(6,6))
plt.imshow(img)
plt.axis('off')
plt.title(f"Predicted: {pred_class}\nConfidence: {confidence*100:.2f}%\nAge: {age}, Sex: {sex}, Location: {location}")
plt.show()

print(f"\nPrediction Result:")
print(f"Class: {pred_class}")
print(f"Confidence: {confidence*100:.2f}%")
print(f"Patient Age: {age}, Sex: {sex}, Lesion Location: {location}")
```
